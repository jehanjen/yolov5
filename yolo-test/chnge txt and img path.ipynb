```python
#change the file path 
import os
import shutil

def save_files(folder_path, txt_path, img_path):
    # Ensure the output directories exist
    os.makedirs(txt_path, exist_ok=True)
    os.makedirs(img_path, exist_ok=True)

    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)

        if os.path.isfile(file_path):
            # Save text files to txt_path
            if filename.lower().endswith('.txt'):
                shutil.copy(file_path, txt_path)

            # Save image files to img_path
            elif any(filename.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png', '.gif']):
                shutil.copy(file_path, img_path)

# Example usage
folder_path = r"C:\Users\jen_bim-y\Desktop\Desktop\yolov5-master\yolo-test\yolosample\ts"  # Specify the path to the folder containing the files
txt_path = r"C:\Users\jen_bim-y\Desktop\Desktop\yolov5-master\yolo_test\yolosliced\txt"  # Specify the path to save the text files
img_path = r"C:\Users\jen_bim-y\Desktop\Desktop\yolov5-master\yolo_test\yolosliced\img"  # Specify the path to save the image files

save_files(folder_path, txt_path, img_path)
```

```python
import os

def remove_pano_txt_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith('pano.txt'):
            filepath = os.path.join(directory, filename)
            os.remove(filepath)
            print(f"Deleted: {filepath}")

# Set the directory containing your text files
directory = r'C:\Users\jen_bim-y\Desktop\Desktop\yolov5-master\yolo_test\yolosliced\txt'

remove_pano_txt_files(directory)
```
